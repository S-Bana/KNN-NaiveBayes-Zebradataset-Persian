# 1 - Function Load Dataset

```python
data = pd.read_csv('w500.csv')
```

```python
data
```

```python
header=list(data)
```

```python
def load_dataframe(i):
    df=pd.read_csv('csv/'+str(i)+'.csv')
    ds=df.sort_values("Number",ascending=False)
    return(ds)
```

```python
def sun_words(df):
    s=df['class']
    c=s[0]
    t=[]
    for i in header:
        x=df[df['Names'].str.match(i)]
        t.append(x[x['class'] == c]['Number'].sum())
    t[0]=c
    return(t)
```

```python
for i in range(1,71):
    df=load_dataframe(i)
    x=sun_words(df)
    data.loc[i] = x
```

```python
data.to_csv('csv/data.csv',index=False)
```
