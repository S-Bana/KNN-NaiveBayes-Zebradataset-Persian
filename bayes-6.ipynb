---

```python
df=pd.read_csv('csv/data.csv')
```

```python
df
```

```python
X = df.loc[:, 'سیاسی':].values
```

```python
y = df.loc[:, 'class'].values
```

---

```python
le = LabelEncoder()
```

```python
y = le.fit_transform(y)
```

---

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=1)
```

---

```python
stdsc = StandardScaler()
```

```python
X_train_std = stdsc.fit_transform(X_train)
```

```python
X_test_std = stdsc.transform(X_test)
```

---

```python
gnb = GaussianNB()
```

```python
gnb.fit(X_train_std, y_train)
```

---

```python
GaussianNB(priors=None)
```

---

```python
y_pred = gnb.predict(X_test_std)
```

# average=(macro,micro,weighted)

```python
print("Accuracy Score : "+str((accuracy_score(y_true=y_test, y_pred=y_pred))*100)+"%")
```

```python
print("F-Score : "+str((f1_score(y_true=y_test, y_pred=y_pred, average='macro'))*100)+"%")
```

```python
print("Recall Score : "+str((recall_score(y_true=y_test, y_pred=y_pred, average='macro'))*100)+"%")
```

```python
print("Precision Score : "+str((precision_score(y_true=y_test, y_pred=y_pred, average='macro'))*100)+"%")
```

---

```python
confusion_matrix(y_true=y_test, y_pred=y_pred)
```

```python
plot_confusion_matrix(gnb,X_test,y_test);
```
