# 1 - Function Load Dataset

```python
def fun_load_dataset():
    li = []
    for i in range(1,8):
        df=pd.read_csv('csv/'+str(i)+'.csv')
        li.append(df)
    frame = pd.concat(li, axis=0, ignore_index=True)
    return(frame)
```

```python
ds=fun_load_dataset()
ds
```

# 2 - Sort Words 

```python
sort_ds=ds.sort_values("Number",ascending=False)
```

```python
sort_ds
```

# 3 - Select 500 words

```python
w_s_ds=sort_ds[["Names"]].iloc[0:500]
```

```python
header=list(w_s_ds['Names'])
```

# 4 - Save 500 words

```python
data = pd.DataFrame(columns=['class']+header)
```

```python
data.to_csv('w500.csv',index=False)
```
