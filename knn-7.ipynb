---

```python
df=pd.read_csv('csv/data.csv')
```

```python
df
```

```python
X = df.loc[:, 'سیاسی':].values
```

```python
y = df.loc[:, 'class'].values
```

---

```python
le = LabelEncoder()
```

```python
y = le.fit_transform(y)
```

---

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=1)
```

---

```python
knn1 = KNeighborsClassifier(n_neighbors=1)
knn5 = KNeighborsClassifier(n_neighbors=5)
knn15 = KNeighborsClassifier(n_neighbors=15)
```

```python
knn1.fit(X_train, y_train)
knn5.fit(X_train, y_train)
knn15.fit(X_train, y_train)
```

```python
y_pred1 = knn1.predict(X_test)
y_pred5 = knn5.predict(X_test)
y_pred15 = knn15.predict(X_test)
```

---

```python
print("Accuracy Score (1) : "+str((accuracy_score(y_true=y_test, y_pred=y_pred1))*100)+"%")
print("Accuracy Score (5) : "+str((accuracy_score(y_true=y_test, y_pred=y_pred5))*100)+"%")
print("Accuracy Score (15) : "+str((accuracy_score(y_true=y_test, y_pred=y_pred15))*100)+"%")
```

```python
confusion_matrix(y_true=y_test, y_pred=y_pred1)
```

```python
confusion_matrix(y_true=y_test, y_pred=y_pred5)
```

```python
confusion_matrix(y_true=y_test, y_pred=y_pred15)
```

---

```python
plot_confusion_matrix(knn1,X_test,y_test);
```

```python
plot_confusion_matrix(knn5,X_test,y_test);
```
