```python
nameList=glob.glob('persian-stopwords/*.sb')
```

```python
w=[]
for i in nameList:
    with open(i,'r') as f:
        x=f.read().splitlines()
        w.extend(x)
```

```python
dframe = pd.DataFrame(w,columns=['words'])
```

```python
dframe.to_csv('inaptW.csv',index=False)
```
