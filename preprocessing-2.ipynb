# 1 - Function List text files

```python
def fun_dir(x):
    nameList=glob.glob(x)
    return(nameList)
```

## 2-1 - Subfunction Load text Files

```python
def openTxt(x):
    with open(x,'r') as f:
        w=f.read()
        w=re.sub(r'[0-9\(\)/]+', ' ', w)
        return(w)
```

## 2-2 - Subfunction ًRgularization Text Files 

```python
def regulTxt(w):
    d=['\u200c','\ufeff','\u200e','\u200f','&nbsp','&amp','nbsp','=','»','«','××','×','(',',','«','•','++','–','»','*','؛','؟',')','.','،','/','<','>','[',']',';',':','!','?','|','_','-','\n']
    
    for j in d:
        w=w.replace(j,"")
        
    normalizer = Normalizer()
    w=normalizer.affix_spacing(w)
    w=normalizer.character_refinement(w)
    
    stemmer = Stemmer()
    w=stemmer.stem(w)
    
    wt=word_tokenize(w)
    
    return(wt)
```

## 2-3 - Subfunction Stopwords

```python
def stopTxt(w):
    p=list(filterfalse(stopwords_list().__contains__, w))

    d=pd.read_csv('inaptW.csv')
    d=list(d['words'])
    p=list(filterfalse(d.__contains__, p))
    return (p)
```

## 2-4 - Subfunction Counter

```python
def contTxt(s):
    a=dict(Counter(s))
    return(a)
```

# 2 - Function Create Words 

```python
def fun_create_words(nameList):
    num=len(nameList)
    i=0
    b=[]
    while(i<num):
        dirList=glob.glob(nameList[i])#nameList[i]+"/*.txt"
        for j in range(len(dirList)):
            z=openTxt(dirList[j])
            w=regulTxt(z)
            s=stopTxt(w)
            b=b+s
        i+=1
    r=contTxt(b)
    return(r)
```
